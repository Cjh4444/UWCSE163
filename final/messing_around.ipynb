```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("spotify_songs.csv")
pd.set_option("display.max_rows", None)
data.columns
```

```python
df = pd.DataFrame(data, columns=["tempo"])
df["tempo"] = df["tempo"].round().astype(int)
```

```python
df["tempo"].value_counts().sort_index().plot(
    kind="bar", xticks=list(range(0, 201, 50))
)
```
