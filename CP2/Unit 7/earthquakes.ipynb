```python
import pandas as pd

df = pd.read_csv('earthquakes.csv')
df.head() # Method to only display the first few rows
```

```python
# Problem 0
ans0 = df["magnitude"].sum() / len(df)
print(ans0)
```

```python
# Problem 1
df1 = df[df["magnitude"] > ans0]
ans1 = len(df1)
print(ans1)
df1.head()
```

```python
# Problem 2
is_tonga = df["name"] == "Tonga"
is_papua_new_guinea = df["name"] == "Papua New Guinea"

tonga_papua_df = df[is_tonga | is_papua_new_guinea]

is_8th_of_month = tonga_papua_df["day"] == 8

ans2 = tonga_papua_df[is_8th_of_month]
ans2.head()
```

```python
# Problem 3
max_idx = df["magnitude"].idxmax()

ans3 = df.iloc[max_idx]
print(ans3)
print(type(ans3))
```
