```python
import pandas as pd

df = pd.read_csv('ufos.csv')
df.head() # Method to only display the first few rows
```

```python
# Problem 

num_changing = len(df[df["shape"] == "changing"])
sum = df[df["shape"] == "changing"]["duration (seconds)"].sum()

print(num_changing)
print(sum)
print(sum / num_changing)


ans0 = df.groupby("shape")["duration (seconds)"].mean()
print("%f" % ans0)
```
