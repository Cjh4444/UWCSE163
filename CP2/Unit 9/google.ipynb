```python
import pandas as pd
import matplotlib.pyplot as plt

# A special magic command to make plots work in the notebook
%matplotlib inline
```

```python
# Problem 0
goog = pd.read_csv("google.csv", index_col="Date", parse_dates=True)
goog.head()
```

```python
# Problem 1
ans1 = goog["Adj Close"]
ans1.plot()

plt.savefig('prob1.png')
```

```python
# Problem 2
ans2 = ans1.loc['2019-10' : '2020-04']
ans2.plot()
plt.savefig('prob2.png')
```

```python
ans3 = ans1.resample("M").mean()
ans3.plot()
plt.savefig('prob3.png')
```

```python
ans4 = ans1.rolling(14).mean()

ans4.plot()
ans4

plt.savefig('prob4.png')
```
