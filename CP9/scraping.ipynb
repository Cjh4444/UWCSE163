```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from bs4.element import NavigableString
```

```python
url = "https://www.w3schools.com/cssref/css_selectors.asp"
page = requests.get(url)
full_of_soup: BeautifulSoup = BeautifulSoup(page.content, "html.parser")
```

```python
table = full_of_soup.find("table")
table
```

```python
rows = table.find_all("tr")[1:]
rows
```

```python
row = rows[0]

selectors: list[str] = []
examples: list[str] = []
descriptions: list[str] = []

for row in rows:
    print(row)
    tds = row.find_all("td")
    selectors.append(tds[0].text)
    examples.append(tds[1].text)
    descriptions.append(tds[2].text)
```

```python
df = pd.DataFrame(
    {"Selector": selectors, "Example": examples, "Description": descriptions}
)
df
```
